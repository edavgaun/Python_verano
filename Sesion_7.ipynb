```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
datos=sns.load_dataset("flights")
datos.head()
```

```
pd.crosstab(datos.month,datos.year)
```

```
datos.pivot("month","year", "passengers")
```

```
sns.heatmap(datos.pivot("month","year", "passengers"))
plt.show()
```

```
sns.heatmap(datos.pivot("month","year", "passengers"), cmap="Reds")
plt.show()
```

```
# https://matplotlib.org/3.2.2/tutorials/colors/colormaps.html

fig, axs =plt.subplots(figsize=(10,7))
sns.heatmap(datos.pivot("month","year", "passengers"), cmap="inferno", ax=axs, annot=True, fmt="d")
plt.show()
```
