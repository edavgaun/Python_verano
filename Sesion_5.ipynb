```python
def mensajes (num):
    t_0=time.time()
    for n in range(num):
        print(n)
    t_1=time.time()
    delta=t_1-t_0
    print(delta)
    return delta
```

```python
mensajes(20000)
```

```python
n=list(range(1,200))
```

```python
tiempos=[]
for valor in n:
    tiempos.append(mensajes(valor))
tiempos
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(n, tiempos)
plt.show()
```

```python
# Creando una cuenta regresiva en python
n=10

for x in range(n):
    print(str(n)+"\t", end="\r")
    n-=1
    time.sleep(1)
print(str(n)+"\t", end="\r")
```

```python
def cuenta_regresiva(timer):
    for x in range(timer):
        print(str(timer)+"\t", end="\r")
        timer-=1
        time.sleep(1)
    print(str(timer)+"\t", end="\r")
```

```python
cuenta_regresiva(20)
```

```python
animacion=["|","/","-","\\"]

for x in range(10):
    for caracter in animacion:
        print(caracter, end="\r")
        time.sleep(0.2)
```

# Código prueba psicométrica

```python
import numpy as np
from termcolor import colored
import time

t_0= time.time() + 30
t_1= 0
aciertos=0
intentos=0
colores=["red", "green", "yellow", "blue", "magenta", "cyan", "white"]

print("Bienvenido a esta actividad psicométrica.\n")

while True:
    intentos+=1
    t_1=time.time()
    tiempo_restante=t_0-t_1
    if tiempo_restante<=0:
        print("-"*40)
        print("El ejercicio ha terminado, gracias")
        print("Número de aciertos",aciertos,"/",intentos)
        break
    while True:
        texto=np.random.choice(colores)
        color=np.random.choice(colores)
        if texto!=color:
            break

    texto_coloreado=colored(texto,color)

    print(tiempo_restante)
    print(texto_coloreado)
    respuesta=input("¿Color?: ")
    respuesta=respuesta.lower()

    if respuesta==color:
        aciertos+=1
```

```python
def psi_test(tiempo):
    import numpy as np
    from termcolor import colored
    import time

    t_0= time.time() + tiempo
    t_1= 0
    aciertos=0
    intentos=0
    colores=["red", "green", "yellow", "blue", "magenta", "cyan", "white"]

    print("Bienvenido a esta actividad psicométrica.\n")

    while True:
        intentos+=1
        t_1=time.time()
        tiempo_restante=t_0-t_1
        if tiempo_restante<=0:
            print("-"*40)
            print("El ejercicio ha terminado, gracias")
            print("Número de aciertos",aciertos,"/",intentos)
            break
        while True:
            texto=np.random.choice(colores)
            color=np.random.choice(colores)
            if texto!=color:
                break

        texto_coloreado=colored(texto,color)

        print("Quedan",round(tiempo_restante,0),"segundos")
        print(texto_coloreado)
        respuesta=input("¿Color?: ")
        respuesta=respuesta.lower()

        if respuesta==color:
            aciertos+=1
```

```python
psi_test(15)
```
