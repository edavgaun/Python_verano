# Definición de funciones

```
def cosulta_pin():
  nip=2020

  while True:
    # Verificando que pin sea numérico
    try:
      pin=int(input("Dame tu pin: "))
      # Verificando que pin sea igual a nip
      if pin!=nip:
        print("Error de pin")
      else:
        print("Acceso concedido!!!")
        break
    except:
      print("Error de valor")
```

```
def edad_nombre():
  # Solicitando la edad:
  while True:
    try:
      edad=int(input("Dame tu edad: "))
      break
    except:
      print("Error, el valor ingresado no corresponde a una edad")
  
  # Solicitando el nombre:
  while True:
    try:
      nombre=input("Dame tu Nombre: ")
      if nombre.isalpha()==True:
        break
      else:
        print("Error en el nombre ingresado")  
    except:
      print("Error en el nombre ingresado")
  
  print("Tu nombre es:", nombre)
  print("Tu edad es:", edad)
```

```
edad_nombre()
```

# Proyecto 1

**Juego Adivinar el número**

```
def juego_aleatorio():
  import numpy as np

  aleatorio=np.random.randint(1,100)
  intentos=0

  while True:
    try:
      numero=int(input("Dame un número entero: "))
      intentos+=1
      if numero==aleatorio:
        print("\n\n\n","-"*40)
        print("Felicidades, has adivinado el número")
        print("El número de intentos fué", intentos)
        break
      elif numero>aleatorio:
        print("El número ingresado es mayor")
      else:
        print("El número ingresado es menor")
    except:
      intentos+=1
      print("El valor ingresado no es un número entero")
```

```
juego_aleatorio()
```
